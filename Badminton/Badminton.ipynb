```python
initial_rating = 1000
k_constant = 30
```

```python
import csv
from pprint import pprint

file = open('/Users/anirudh-15229/Desktop/Anirudh Personal/Badminton-ELO-Rankings/Badminton/Badminton 3.0.csv')
csvreader = csv.reader(file)
header = []
header = next(csvreader)
rows = []
```

```python
class Player:
  def __init__(self, name, debut):
    self.name = name
    self.debut = debut
    self.rating = initial_rating
    self.wins = 0
    self.losses = 0
    self.matches = 0
    self.current_win_streak = 0
    self.current_loss_streak = 0
    self.highest_win_streak = 0
    self.highest_loss_streak = 0
    self.highest_rating = initial_rating
    self.lowest_rating = initial_rating
    self.win_against_counter = {}
    self.lose_against_counter = {}
    self.win_with_counter = {}
    self.lose_with_counter = {}

  def print(self):
   print(self.__dict__)
   print()
  #  print("Name : ",self.name,",Debut : ",self.debut,",Rating : ",self.rating,", Wins : ",self.wins,", Losses : ",self.losses,",Matches : ",self.matches,",Current Win Streak : ",self.current_win_streak,",Current Loss Streak : ",self.current_loss_streak,",Highest Win Streak : ",self.highest_win_streak,",Highest Loss Streak : ",self.highest_loss_streak,", Win % : ", (self.wins * 100)/(self.wins + self.losses),", Highest rating : ",self.highest_rating,",Lowest rating : ",self.lowest_rating)
players = {}
```

```python
max_points_gained=0
max_date=""
max_winner_1 = ""
max_winner_2 = ""
max_loser_1 = ""
max_loser_2 = ""
for row in csvreader:
    date = row[0]
    winner_1 = row[1]
    winner_2 = row[2]
    loser_1 = row[3]
    loser_2 = row[4]


    if winner_1 not in players.keys():
        players[winner_1] = Player(winner_1,date)

    if winner_2 not in players.keys():
        players[winner_2] = Player(winner_2,date)
    
    if loser_1 not in players.keys():
        players[loser_1] = Player(loser_1,date)

    if loser_2 not in players.keys():
        players[loser_2] = Player(loser_2,date)

    team_1_rating = (players[winner_1].rating + players[winner_2].rating) / 2
    team_2_rating = (players[loser_1].rating + players[loser_2].rating) / 2

    p1 = 1.0 / (1.0 + (10 ** ((team_1_rating - team_2_rating)/400)))
    p2 = 1.0 / (1.0 + (10 ** ((team_2_rating - team_1_rating)/400)))

    #Team 1 won, Team 2 lost
    if k_constant * (1 - p2) > max_points_gained :
        max_points_gained = k_constant * (1 - p2)
        max_date= date
        max_winner_1=winner_1
        max_winner_2=winner_2
        max_loser_1=loser_1
        max_loser_2=loser_2
        
    players[winner_1].rating = players[winner_1].rating + k_constant * (1 - p2)
    players[winner_1].wins = players[winner_1].wins + 1
    players[winner_1].matches = players[winner_1].matches + 1
    players[winner_1].current_win_streak = players[winner_1].current_win_streak + 1
    players[winner_1].current_loss_streak = 0

    players[winner_1].win_against_counter[loser_1] = players[winner_1].win_against_counter.get(loser_1,0) + 1
    players[winner_1].win_against_counter[loser_2] = players[winner_1].win_against_counter.get(loser_2,0) + 1
    players[winner_1].win_with_counter[winner_2] = players[winner_1].win_with_counter.get(winner_2,0) + 1
    
    if players[winner_1].rating > players[winner_1].highest_rating:
        players[winner_1].highest_rating = players[winner_1].rating
    
    if players[winner_1].current_win_streak > players[winner_1].highest_win_streak:
        players[winner_1].highest_win_streak = players[winner_1].current_win_streak


    players[winner_2].rating = players[winner_2].rating + k_constant * (1 - p2)
    players[winner_2].wins = players[winner_2].wins + 1
    players[winner_2].matches = players[winner_2].matches + 1
    players[winner_2].current_win_streak = players[winner_2].current_win_streak + 1
    players[winner_2].current_loss_streak = 0

    players[winner_2].win_against_counter[loser_1] = players[winner_2].win_against_counter.get(loser_1,0) + 1
    players[winner_2].win_against_counter[loser_2] = players[winner_2].win_against_counter.get(loser_2,0) + 1
    players[winner_2].win_with_counter[winner_1] = players[winner_2].win_with_counter.get(winner_1,0) + 1

    if players[winner_2].rating > players[winner_2].highest_rating:
        players[winner_2].highest_rating = players[winner_2].rating

    if players[winner_2].current_win_streak > players[winner_2].highest_win_streak:
        players[winner_2].highest_win_streak = players[winner_2].current_win_streak


    players[loser_1].rating = players[loser_1].rating + k_constant * (0 - p1)
    players[loser_1].losses = players[loser_1].losses + 1
    players[loser_1].matches = players[loser_1].matches + 1
    players[loser_1].current_loss_streak = players[loser_1].current_loss_streak + 1
    players[loser_1].current_win_streak = 0

    players[loser_1].lose_against_counter[winner_1] = players[loser_1].lose_against_counter.get(winner_1,0) + 1
    players[loser_1].lose_against_counter[winner_2] = players[loser_1].lose_against_counter.get(winner_2,0) + 1
    players[loser_1].lose_with_counter[loser_2] = players[loser_1].lose_with_counter.get(loser_2,0) + 1

    if players[loser_1].rating < players[loser_1].lowest_rating:
        players[loser_1].lowest_rating = players[loser_1].rating

    if players[loser_1].current_loss_streak > players[loser_1].highest_loss_streak:
        players[loser_1].highest_loss_streak = players[loser_1].current_loss_streak


    players[loser_2].rating = players[loser_2].rating + k_constant * (0 - p1)
    players[loser_2].losses = players[loser_2].losses + 1
    players[loser_2].matches = players[loser_2].matches + 1
    players[loser_2].current_loss_streak = players[loser_2].current_loss_streak + 1
    players[loser_2].current_win_streak = 0

    players[loser_2].lose_against_counter[winner_1] = players[loser_2].lose_against_counter.get(winner_1,0) + 1
    players[loser_2].lose_against_counter[winner_2] = players[loser_2].lose_against_counter.get(winner_2,0) + 1
    players[loser_2].lose_with_counter[loser_1] = players[loser_2].lose_with_counter.get(loser_1,0) + 1

    if players[loser_2].rating < players[loser_2].lowest_rating:
        players[loser_2].lowest_rating = players[loser_2].rating

    if players[loser_2].current_loss_streak > players[loser_2].highest_loss_streak:
        players[loser_2].highest_loss_streak = players[loser_2].current_loss_streak

print(max_points_gained,max_date,max_winner_1,max_winner_2,max_loser_1,max_loser_2)
```
